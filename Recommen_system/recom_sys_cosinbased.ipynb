```python
# count the number of duplicates and add the count in front of each value
Liked_songs['repeated'] = Liked_songs.groupby('id')['id'].transform('size')

# drop the duplicates
Liked_songs = Liked_songs.drop_duplicates(subset='id')

# sort the values by count in descending order
Liked_songs = Liked_songs.sort_values(by='repeated', ascending=False)

# reset the index
Liked_songs = Liked_songs.reset_index(drop=True)

# display the final result
Liked_songs

## Adding The liked songs to the database as 1 (Liked) 0(not liked)
liked_ids = set(Liked_songs['id'])
database['liked'] = database['id'].isin(liked_ids).astype(int)
```

```python
#Our dataset is soo big let's make it smaller 

sample = database.sample(n=100000, replace=False)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer(stop_words='english')
sample['name'] = sample['name'].fillna('')
tfidf_matrix = tfidf.fit_transform(sample['name'])
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = database.loc[database['name'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    song_indices = [i[0] for i in sim_scores]
    return database['name'].iloc[song_indices]
```

```python
get_recommendations("L'italiano")
```

```python
all_recoms = []
for i in Liked_songs["name"]:
    try: 
       all_recoms.append(get_recommendations(i))
    except:
        continue
```

```python
songs = np.unique(all_recoms)
recommended_songs = pd.DataFrame(songs, columns = ['Musics'])

recommended_songs.to_csv("recommended_songs.csv")
```
